### Fasitløsning og krav til godkjenning
- To løsningsalternativer
    - nøstede løkker
    - en løkker som kaller en funksjon som igjen inneholder en løkke
- Det må tegnes en spiral as sekskanter
- Det skal være 20 sekskanter
- Oppgaven godkjennes også hvis den minste sekskanten har sidelengde 0

```python
# variant 1
import turtle
turtle.clearscreen()
sam = turtle.Turtle()
for n in range(20):
    for s in range(6):
        sam.forward((n+1) * 10)
        sam.left(60)
    sam.left(13)
```

```python
# variant 2
import turtle

def sekskant(padde, lengde):
    for s in range(6):
        padde.forward(lengde)
        padde.left(60)

turtle.clearscreen()
sam = turtle.Turtle()
for n in range(20):
    sekskant(sam, (n+1) * 10)
    sam.left(13)
```

### Fasitløsning og krav til godkjenning
1. Kurven for $\sin x^2$ skal være glatt, dvs tegnet med nok punkter
1. Det skal være tittel, benevnelse på x- og y-aksen og legende
1. Grensene skal være satt til $0-2\pi$ (cirka) langs x-aksen
1. Oppgaven godkjennes også hvis ett av elementene under nr 2 og 3 mangler

```python
# her kommer løsningen

x = linspace(0, 2*pi, 250)   # flere punkter langs x-aksen
y = sin(x)*sin(x)
z = sin(x*x)
plot(x, y, label='sin(x)*sin(x)')  # label
plot(x, z, label='sin(x*x)')       # label

xlim(0, 2*pi)
ylim(-1.05, 1.05)
xlabel('x')
ylabel('y(x)')
legend()
title('Kvadrat av sinus vs sinus av kvadrat');
```

```python
# Denne løsningen er ennå litt finere
# Legenden er flyttet til nedre venstre hjørne
# Legenden og aksene benytter LaTeX

x = linspace(0, 2*pi, 250)   # flere punkter langs x-aksen
y = sin(x)*sin(x)
z = sin(x*x)
plot(x, y, label='$\sin^2 x$')  # label
plot(x, z, label='$\sin\,x^2$')       # label

xlim(0, 2*pi)
ylim(-1.05, 1.05)
xlabel('$x$')
ylabel('$y(x$)')
legend(loc='best')
title('Kvadrat av sinus vs sinus av kvadrat');
```

Du skal i denne oppgaven benytte den idéen til å implementere et Lindenmeyer-System, gjerne snøfnuggen. Den lange `else` delen i `snowflake()` funksjonen, som implementerer regelen, krymper da til en liten for-løkke som i eksemplet ovenfor.

### Råd
- Du må definere en funksjon per operasjon, dvs `F`, `+` og `-`. 
- Funksjonene for `+` og `-` tregner ikke `depth` informasjon, men `F` gjør det. Siden alle funksjoner skal kalles på samme måte, må også `+` og `-` funksjoner kunne ta imot `depth`, selv om de ignorerer denne informasjonen.

### Fasitløsning og krav til godkjenning

- Det finnes to alternative løsninger under
- Koden må virke og tegne et snøfnugg (minst en side)
- Rekursjonen skal utføres ved å iterere over symbolene i regelen og kalle de riktige operasjonene fra dictionary

```python
# i denne løsningen sendes dict med anvisninger med til alle funksjoner

def forward(padde, depth, operations):
    if depth == 0:
        padde.forward(10)
    else:
        for symbol in 'F+F--F+F':
            operations[symbol](padde, depth-1, operations)

def right(padde, depth, operations):
    padde.right(60)
    
def left(padde, depth, operations):
    padde.left(60)

def snowflake(padde, depth):  # denne funksjonen er representerer startsymbolet
    
    snowflake_ops = {'F': forward, '+': left, '-': right}

    snowflake_ops['F'](padde, depth, snowflake_ops)  # F er startsymbol
    

turtle.clearscreen()
sam = turtle.Turtle()

snowflake(sam, 3)
```

```python
# i denne løsningen brukes en global variable med operasjonene, det er også ok

def forward(padde, depth):
    if depth == 0:
        padde.forward(10)
    else:
        for symbol in 'F+F--F+F':
            operations[symbol](padde, depth-1)

def right(padde, depth):
    padde.right(60)
    
def left(padde, depth):
    padde.left(60)

operations = {'F': forward, '+': left, '-': right}

def snowflake(padde, depth):
    operations['F'](padde, depth)  
    

turtle.clearscreen()
sam = turtle.Turtle()

snowflake(sam, 3)
```

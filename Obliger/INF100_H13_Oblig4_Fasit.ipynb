## Oppgave 12: Middeltemperaturer på måned ##

I denne oppgaven skal du beregne og tegne opp middels temperatur for hver måned i 2012 for værstasjonen på Ås.

1. Tar utgangspunkt i hvordan vi beregnet og plottet middels temperatur per uke i forelesningen 17.10.
1. Vi deler året inn i 12 måneder à 30 dager og ser bort fra de siste 6 dagene i året.
1. Du må altså bruke de første 360 elementene i `T_avg` og omforme dem til en array med 12 rader og 30 kolonner.
1. Så beregner du middelverdiene for hver av de 12 månedene.
1. Månedsverdiene skal plottes sammen med døgnverdiene, tilsvarende grafikken fra forelesningen vist nedenfor. Månedsverdiene skal merkes med symboler.

**NB: Du må laste ned filen `weather_umb_2012.csv` til samme mappe som denne notebook-filen.**

Hvis du vil ha en liten ekstra utfordring, utforsk funksjonen `step()` som et alternativ til `plot()` for å tegne månedsverdiene.

```python
from IPython.display import Image ; Image(filename='Figures/O12_sample.png')
```

### Fasit og krav til godkjenning

- Månedsmidlene skal være beregnet korrekt
- Det skal være legende og akseoppmerking
- Døgn- og månedsmidlene skal være med i grafen

```python
# Arbeide i denne cellen!

weather = loadtxt('weather_umb_2012.csv', delimiter=';', usecols=(1, 2, 3, 4, 5))
day = linspace(1, 366, num=366)

T_avg = weather[:, 0]
T_avg_by_month = T_avg[:360].reshape(12, 30)
T_avg_month_mean = T_avg_by_month.mean(axis=1)
len(T_avg_month_mean)


x_month = 15 + 30 * linspace(0, 11, num=12)
plot(day, T_avg, '.', label=u'Døgngjennomsnitt');
plot(x_month, T_avg_month_mean, 'o-', linewidth=3, label=u'Månedsgjennomsnitt');
xlabel(u'Døgn regnet fra nyttår');
ylabel(u'Temperatur [C]');
title(u'Temperatur for Ås i 2012');
xlim(0, 367);
ylim(-20, 25);
legend(loc='lower center');
```

#### Alternativ løsning med `step()`

```python
weather = loadtxt('weather_umb_2012.csv', delimiter=';', usecols=(1, 2, 3, 4, 5))
day = linspace(1, 366, num=366)
T_avg = weather[:, 0]
T_avg_by_month = T_avg[:360].reshape(12, 30)
T_avg_month_mean = T_avg_by_month.mean(axis=1)
len(T_avg_month_mean)


x_month = 15 + 30 * linspace(0, 11, num=12)
plot(day, T_avg, '.', label=u'Døgngjennomsnitt');
step(x_month, T_avg_month_mean, where='mid', linewidth=3, label=u'Månedsgjennomsnitt');
xlabel(u'Døgn regnet fra nyttår');
ylabel(u'Temperatur [C]');
title(u'Temperatur for Ås i 2012');
xlim(0, 367);
ylim(-20, 25);
legend(loc='lower center');
```

## Bonusoppgave 4 ##

Du skal lage en noe mer avanset variant av grafikken over ukegjennomsnittstemperatur fra forelesningen (jf eksempelgrafikken ovenfor). Du kan ta utgangpunkt i koden fra forelesningen.

1. I tillegg til å beregne middelverdien for temperaturen for hver uke, skal du også beregne standardavviket. Dette gjøres med kommand `std()` som brukes på akkurat samme måte som `mean()`.
1. I grafikken skal du legge lys grønn bakgrunn mellom middelverdi minus standardavvik og middelverdi pluss standardavvik. Til dette kan du bruke funksjonen `fill_between()`. Se på dokumentasjon for `fill_between()` for hvordan du bruker den.

Den resulterende grafikken skal se omtrent slik ut:

```python
from IPython.display import Image ; Image(filename='Figures/B4_sample.png')
```

### Fasit og krav til godkjenning

- std.dev må være beregnet korrekt
- fill_between må brukes riktig
- kurven skal være merket opp 

```python
weather = loadtxt('weather_umb_2012.csv', delimiter=';', usecols=(1, 2, 3, 4, 5))
day = linspace(1, 366, num=366)
T_avg = weather[:, 0]
T_avg_by_week = T_avg[:364].reshape(52, 7)
mn = T_avg_by_week.mean(axis=1)
sd = T_avg_by_week.std(axis=1)
x_week = 3.5 + 7 * linspace(0, 51, num=52)
plot(day, T_avg, '.', label=u'Døgngjennomsnitt');
plot(x_week, mn, '-', linewidth=1, label=u'Ukesgjennomsnitt');
fill_between(x_week, mn-sd, mn+sd, color='g', alpha=0.4, label=u'+- std.avvik');
xlabel(u'Døgn regnet fra nyttår');
ylabel(u'Temperatur [C]');
title(u'Temperatur for Ås i 2012');
xlim(0, 367);
ylim(-20, 25);
legend(loc='lower center');
```

```python
import turtle

sam = turtle.Turtle()

sam.width(5)
sam.hideturtle()
sam.speed('fastest')

sam.up()
sam.goto(-300, -250)
sam.down()
sam.color('blue')

# H
sam.setheading(90)
sam.forward(100)
sam.backward(50)
sam.right(90)
sam.forward(80)
sam.left(90)
sam.forward(50)
sam.backward(100)
sam.right(90)

# A
sam.up()
sam.forward(20)
sam.down()

sam.left(90)
sam.forward(100)
sam.right(90)
sam.forward(80)
sam.backward(80)
sam.left(90)
sam.backward(50)
sam.right(90)
sam.forward(80)
sam.left(90)
sam.forward(50)
sam.backward(100)
sam.right(90)

# N
sam.up()
sam.forward(20)
sam.down()

sam.left(90)
sam.forward(100)
x, y = sam.position()
sam.goto(x+80, y-100)
sam.forward(100)
sam.backward(100)
sam.right(90)

# S
sam.up()
sam.forward(20)
sam.down()

sam.setheading(0)
sam.forward(80)
sam.left(90)
sam.forward(50)
sam.left(90)
sam.forward(80)
sam.right(90)
sam.forward(50)
sam.right(90)
sam.forward(80)
```

## Bonusoppgave 1 ##

Skriv en Python funksjon som tar imot en streng og tegner teksten, dvs `draw_text('HANS')` skal tegne `HANS` som i eksempelet for oppgave 2. Du kan få godkjent oppgaven hvis koden din støtter minst 10 forskjellige bokstaver eller tegn.

```python
import turtle

def skriv(tekst):
    """Funksjonen skriver teksten ved hjelp av Turtle Graphics."""
    
    # forberedelser
    kalligraf = turtle.Turtle()
    kalligraf.hideturtle()
    kalligraf.speed('fastest')
    kalligraf.width(5)
    kalligraf.up()
    kalligraf.goto(-350, 0)
    
    # skriv teksten
    for tegn in tekst:
        skriv_bokstav(kalligraf, tegn)
        
def skriv_bokstav(padde, bokstav):
    """Funksjonen skriver en bokstav og flytter til startpunkt for neste bokstav."""
    
    # husk posisjon i starten og sikre at pennen er nede
    x, y = padde.position()
    padde.down()
    
    # skriv selve bokstaven
    if bokstav == 'A':
        bredde = skriv_A(padde)
    elif bokstav == 'H':
        bredde = skriv_H(padde)
    elif bokstav == 'I':
        bredde = skriv_I(padde)
    elif bokstav == 'N':
        bredde = skriv_N(padde)
    elif bokstav == 'S':
        bredde = skriv_S(padde)
    elif bokstav == ' ':
        bredde = 20   # mellomrom, ingenting å tegne
    else:
        bredde = skriv_strek(padde)
        
    # flytt padden til startposisjon for neste bokstav
    padde.up()
    padde.goto(x+bredde+10, y)
    
def skriv_A(padde):
    """Skriver ut A og returnerer bokstavens bredde."""
    
    padde.setheading(90)
    padde.forward(100)
    padde.right(90)
    padde.forward(80)
    padde.backward(80)
    padde.left(90)
    padde.backward(50)
    padde.right(90)
    padde.forward(80)
    padde.left(90)
    padde.forward(50)
    padde.backward(100)

    return 80

def skriv_H(padde):
    """Skriver ut H og returnerer bokstavens bredde."""
    
    padde.setheading(90)
    padde.forward(100)
    padde.backward(50)
    padde.right(90)
    padde.forward(80)
    padde.left(90)
    padde.forward(50)
    padde.backward(100)

    return 80

def skriv_I(padde):
    """Skriver ut I og returnerer bokstavens bredde."""
    
    padde.setheading(90)
    padde.forward(100)
    
    return 10

def skriv_N(padde):
    """Skriver ut N og returnerer bokstavens bredde."""
    
    padde.setheading(90)
    padde.forward(100)
    x, y = padde.position()
    padde.goto(x+80, y-100)
    padde.forward(100)

    return 80

def skriv_S(padde):
    """Skriver ut S og returnerer bokstavens bredde."""
    
    padde.setheading(0)
    padde.forward(80)
    padde.left(90)
    padde.forward(50)
    padde.left(90)
    padde.forward(80)
    padde.right(90)
    padde.forward(50)
    padde.right(90)
    padde.forward(80)

    return 80

def skriv_strek(padde):
    """Skriver ut _ og returnerer bokstavens bredde."""
    
    padde.setheading(0)
    padde.forward(40)
    
    return 40

# nå benytt dette til å skrive en tekst
turtle.clearscreen()
skriv("HANS SIN SANS")
```

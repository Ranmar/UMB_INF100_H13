## Numerisk matematikk: et eksempel

Det finnes en god del matematiske problemer som vi kan løse med papir og blyant (og den del erfaring, evt [Dragonbox](http://www.dragonboxapp.com/) trening). Hvis vi f eks har

$$
2 x + 3 = 5
$$

finner vi

$$
x = \frac{5 - 3}{2} = 1
$$

som løsning. Er kvadrater involvert, blir det litt vanskeligere og vi vil få to løsninger. F eks

$$
x^2 - 2 = 7
$$

finner vi løsningen

$$
x_{1,2} = \pm \sqrt{7+2} = \pm\sqrt{9} = \pm 3 \;.
$$

For tredje potens, f eks

$$
x^3 -2 x^2 + x - 5 = 9
$$

er det fortsatt mulig å finne en løsning med papir og blyant (f eks med [Cardanos metode](http://en.wikipedia.org/wiki/Cubic_function)), men dette er krevende.  Fra femte potens og oppover finnes ikke lenger noen generelle løsninger som kan finnes med papir og blyant.

Andre tilsynelatende enkle likninger kan ikke løses med papir og blyant i det hele tatt. Skal vi f eks finne den verdien av $x$ som oppfyller

$$
x e^x = 1
$$

så finnes ikke noen papir-og-blyant løsning.

Vi må ty til *numeriske metoder* for å løse likningene.

### Vi plotter likningene

I alle likningene vi har sett på, har vi en funksjon av $x$ på venstre siden og en konstant på høyre siden av likhetstegnet. Vi kan altså skrive alle likningene som

$$
f(x) = c
$$

der $f(x)$ er en (matematisk) funksjon av $x$ og $c$ er en konstant.

Vi anskueliggjør likningene ved å plotte funksjonen $f(x)$ og konstanten $c$ i samme grafikk. 

Vi lager en tegning per likning ovenfor:

- $f(x)$ tegnes ved å beregne funksjonens verdi på mange støttepunkter $x$
- $c$ tegnes som en horizontal linje med $y$-verdi $c$
- $f(x)=c$ er oppfylt der linjene krysses

#### Likning 1: $2x+3=5$

- $f(x) = 2x+3$
- $c=5$

```python
x = linspace(-5, 5)
y = 2 * x + 3
plot(x, y, '-', lw=3, label='$2x+3$')
plot([-5, 5], [5, 5], '-', lw=3, label='$5$')
xlabel('x')
ylabel('y')
legend(loc='best')
grid(True)
title('Likning 1');
```

#### Likning 2: $x^2-2=7$

- $fx) = x^2-2$
- $c=7$

```python
x = linspace(-5, 5)
y = x**2 - 2
plot(x, y, '-', lw=3, label='$x^2-2$')
plot([-5, 5], [7, 7], '-', lw=3, label='$7$')
xlabel('x')
ylabel('y')
legend(loc='best')
grid(True)
title('Likning 2');
```

#### Likning 3: $x^3 -2 x^2 + x - 5 = 9$

- $f(x) = x^3 -2 x^2 + x - 5 = 9$
- $c = 9$

```python
x = linspace(-5, 5)
y = x**3 - 2 * x**2 + x - 5
plot(x, y, '-', lw=3, label='$x^3 -2 x^2 + x - 5$')
plot([-5, 5], [7, 7], '-', lw=3, label='$9$')
xlabel('x')
ylabel('y')
legend(loc='best')
grid(True)
title('Likning 3');
```

#### Likning 4: $x e^x=1$

- $f(x) = x e^x$
- $c = 1$

```python
x = linspace(-5, 5)
y = x * e**x
plot(x, y, '-', lw=3, label='$x e^x$')
plot([-5, 5], [7, 7], '-', lw=3, label='$1$')
xlabel('x')
ylabel('y')
ylim(-1, 10)
legend(loc='best')
grid(True)
title('Likning 4');
```

### Hvordan kan vi finne en løsningen for likningene?

- Hva betyr det "å finne en løsning"?
    - Vi skal finne den verdien av $x$ det blå og grønn linje krysses.
- Mulige problemer
    - Kanskje linjene krysser hverandre aldri?
    - Kanskje linjene krysser hverandre flere ganger (jf Likning 2)?
    - Hvor skal vi lete, mellom $-\inf$ og $+\inf$?
- Vår løsning for nå:
    - Vi skal bare lete innenfor $a \leq x \leq b$
    - Vi forutsetter at linjene krysser hverandere *nøyaktig en gang* mellom $a$ og $b$ (vi sjekker dette på forhånd i grafikkene over)
- Bedre svar gis i matematikken

```python
# Denne cellen definerer en funksjon som jeg skal bruke for å illustrere løsningsmetoden.
# Se gjerne på funksjonen for å se hvordan man kan lage illustrasjoner med. 
#
# Koden i denne cellen inngår ikke i pensumet.

def bisect_plot(func, const, a, b, xmin, xmax, lower, left):
    """Illustrende grafikk for biseksjon i [a, b]."""
    
    # midtpunkt
    m = (a + b) / 2.
    
    # funksjon over hele plotinterval
    x = linspace(xmin, xmax)
    y = func(x)
    
    # array med x-verdier bare for halvdelen som inneholder krysningen
    if (func(a)-const) * (func(m)-const) < 0:
        x_next = linspace(a, m)
    else:
        x_next = linspace(m, b)
    y_next = func(x_next)
    
    # tegn funksjon og konstant
    plot(x, y, '-', lw=3)
    plot([xmin, xmax], [const, const], '-', lw=3)
    
    # set/finn grenser
    xlim(xmin, xmax)
    ymin, ymax = gca().get_ylim()
    
    # fyll areal i intervallet som skal ses på i neste skritt med farge
    fill_between(x_next, y_next, ymin, color=(0.8, 1.0, 0.8))
    
    # tegn stiplede linjer ved a, m, b og f(a), f(m), f(b)
    for val, lbl in ((a, 'a'), (m, 'm'), (b, 'b')):
        plot([val, val, xmin], [ymin, func(val), func(val)], 'k--', lw=3)
        text(val, ymin-lower, '${}$'.format(lbl), fontsize=20,
             horizontalalignment='center',
             verticalalignment='top')
        text(xmin-left, f(val), '$f({})$'.format(lbl), fontsize=20,
             horizontalalignment='right',
             verticalalignment='center')
        text(xmin-left, const, '$c$'.format(lbl), fontsize=20,
             horizontalalignment='right',
             verticalalignment='center')
        
    # text() endrer de vertikale grensene, vi må sette de tilbake
    ylim(ymin, ymax)
```

#### Fremgangsmåte: Biseksjonmetoden ("halvdelingsmetoden", [på Wikipedia](http://en.wikipedia.org/wiki/Bisection_method), [ved Wikiversity](http://en.wikiversity.org/wiki/The_bisection_method))

- Vi tegner først en grafikk til illustrasjon

```python
def f(x):
    return x**2 - 2

bisect_plot(func=f, const=7, a=1, b=4, xmin=-1, xmax=5, lower=2, left=0.5)
```

- Idé:
    1. La oss kalle $x$-verdien der linjene krysses $x^*$. Vi vet at den ligger mellom $a$ og $b$.
    1. $m = (a + b) / 2$ ligger midt imellom $a$ og $b$.
    1. **Enten** ligger $x^*$ til venstre for $m$, **eller** til høyre for $m$
    1. $\Longrightarrow$ $x^*$ ligger altså enten i $[a, m]$ eller i $[m, b]$
    1. Vi velger det intervallet som inneholder $x^*$ og halverer det intervallet.
- Dette illustreres i de følgende grafikkene
- Vi starter fra grafikken over, der vi hadde $a=1$, $b=4$, so $m=2.5$; vi velger intervallet $[m, b]$

```python
bisect_plot(func=f, const=7, a=2.5, b=4, xmin=-1, xmax=5, lower=2, left=0.5)
```

- I denne grafikken er $a=2.5$, $b=4$, $m=3.25$ og krysningen ligger til venstre for $m$, så vi velger intervallet $[a, m]$

```python
bisect_plot(func=f, const=7, a=2.5, b=3.25, xmin=-1, xmax=5, lower=2, left=0.5)
```

- Nå er $a=2.5$, $b=3.25$, $m=2.875$ og neste intervallet blir $[m, b]$.

```python
bisect_plot(func=f, const=7, a=2.875, b=3.25, xmin=-1, xmax=5, lower=2, left=0.5)
```

- Legg merke til at intervallene blir smalere og smalere.
- Intervalbredden *halveres* for hver gang.
- $f(m)$ ligger stadig nærmer $c$
- Når differansen $|f(m)-c|$ er blitt små nok, godtar vi $m$ som **approksimasjon** (tilnærming) til den sanne løsningen $x^*$
- I grafikken over har vi $m=3.0625$ og $f(m)=7.3789$, dvs $|f(m)-c|=0.3789$. Hvis vi er fornøyd med en slik grov løsning, kan vi godta $x^*\approx 3.0625$ som løsning.

#### Numeriske metoder gir tilnærmede løsninger

Dette er en sentral egenskap ved numeriske metoder: de gir løsninger som er *tilnærmet riktig*, dvs *approksimasjoner*, ikke eksakte løsninger. Numerisk matematikk handler i stor grad om

- hvor god approksimasjonene er ved bruk av bestemte metoder
- i hvilke situasjoner man kan benytte (eller ikke må benytte) bestemte metoder
- hvor effektiv bestemte metoder er for visse typer av problemer

### Biseksjon i Python

- Vi gjør først en forenkling og krever at konstanten $c=0$.
    - Dette gjør løsningen enklere.
    - Samtidig er det ingen begrensing siden
    
      $$f(x) = c$$
      $$\Leftrightarrow f(x)-c = 0$$
      $$\Leftrightarrow g(x) = 0$$
      
      med $g(x) = f(x)-c$
      
- Hvordan kan vi finne ut på hvilken side av $m$ løsningen ligger?
    1. $g(x)$ må være mindre enn $c=0$ i den ene intervallenden, større i den andre
    1. $\Longrightarrow$ $g(x)$ må ha forskjellig fortegn i intervalendene
    1. $\Longrightarrow$ $x^*$ ligger i $[a, m]$ hvis $g(a)g(m) < 0$ 
- Nå kan vi gå frem slik:
    1. Start med gitt $a$ og $b$
    1. Repeter så følgende skritt:
        1. Beregn $m=(a+b)/2$
        1. Beregn $g(m)$
        1. Hvis $|g(m)| \leq \epsilon \Longrightarrow$ ferdig, lever $m$ som resultat
        1. Ellers 
            1. Hvis $g(a)g(m) < 0$
                1. Sett $b = m$
            1. Ellers
                1. Sett $a = m$
- Den matematiske funksjonen vi skal benytte må vi definere som Python funksjon
- Da blir det følgende kode

```python
# definer funksjon og konstanten
def g(x):
    return x**2 - 9

epsilon = 0.1    # toleranse

# samme eksempel som ovenfor
a = 1
b = 4

while True:     # en løkke som vil gå for altid

    m = ( a + b ) / 2.0
    gm = g(m)

    # til vår informasjon skriver vi ut a, b, m, g(m) for hver repetisjon
    print a, b, m, gm
    
    if abs(gm) <= epsilon:
        break   # hopp ut av løkken, m er løsningen
    else:
        if g(a) * g(m) < 0:
            b = m
        else:
            a = m    
            
# Nå kan vi skrive resultatet:
print
print "Krysningen er omtrent ved x =", m
print "Der er g(x) =", g(m), "mens ved det sanne krysningspunktet er g(x) = 0."
```

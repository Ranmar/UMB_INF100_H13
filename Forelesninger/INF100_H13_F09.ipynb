## Renteberegning

Denne koden beregner utviklingen av et gitt pengebeløp satt inn på et sparekonto med fast rentesats. Beregningen tar hensyn til at det påløper 28% inntektsskatt på renter. Formueskatt beregnes ikke.

#### Parameter for beregningen

I den påfølgende cellen kan du oppgi beløpet du vil sette inn på konto, rentesats, skattesats og antall år beregningen skal kjøres for.

```python
belop = 100000
antall_aar = 10

rente_i_prosent = 3.0
inntektsskatt_i_prosent = 28.0

# Beregningen utføres av koden som følger. Du behøver ikke å endre noe nedenfor denne linjen.
print u'{:>3s}  {:>10s}      {:>10s}     {:>10s}'.format(u'År', u'Renter', u'Skatt', u'Beløp') 
                                                             
for aar in range(antall_aar):
    
    rente = rente_i_prosent / 100 * belop
    skatt = inntektsskatt_i_prosent / 100 * rente
    
    belop = belop + rente - skatt
    
    print '{:3d}  {:10.2f} kr   {:10.2f} kr  {:10.2f} kr'.format(aar+1, rente, skatt, belop)
```

## Fargesetting i Arduino-kretsen [ikke pensum]

- Sensoren leverer en måleverdi mellom 0 og 1023
- Må konverteres til lysintensitet 0 til 255 for hver av fargene rødt, grønt og blått
- Sensorverdi: $s$
- Lysintensitetene: $r$, $g$, $b$

### Likningene for transformasjon

$$r= \begin{cases} 
          255 \times \frac{s}{341} & 0\leq s < 341 \\[2ex]
          255 \times \frac{682 - s}{341} & 341 \leq s < 682 \\[2ex]
          0 & 682 \leq s \leq 1023
          \end{cases}
          $$
          
* * *          

$$g= \begin{cases} 
          0 & 0 \leq s < 341 \\[2ex]
          255 \times \frac{s-341}{341} & 341\leq s < 682 \\[2ex]
          255 \times \frac{1023 - s}{341} & 682 \leq s \leq 1023
          \end{cases}
          $$


* * *

$$b= \begin{cases} 
          255 \times \frac{341 - s}{341} & 0\leq s < 341 \\[2ex]
          0 & 341\leq s < 682 \\[2ex]
          255 \times \frac{s-682}{341} & 682 \leq s \leq 1023
          \end{cases}
          $$


### Grafisk fremstilling av transformasjonen

```python
sensor_vals = linspace(0, 1023, num=1024)

r_vals = [        (255 * s) / 341 if s < 341 else (255 * (682 - s))/341 if s < 682 else                      0 for s in sensor_vals]
g_vals = [                      0 if s < 341 else (255 * (s - 341))/341 if s < 682 else (255 * (1023 - s))/341 for s in sensor_vals]
b_vals = [(255 * (341 - s)) / 341 if s < 341 else                     0 if s < 682 else (255 * ( s - 682))/341 for s in sensor_vals]

plot(sensor_vals, r_vals, 'r', lw=3)
plot(sensor_vals, g_vals, 'g', lw=3)
plot(sensor_vals, b_vals, 'b', lw=3)
xlim(0, 1023)
ylim(0, 256)
xlabel(u'Sensorens måleverdi $s$')
ylabel(u'Lysintensitet')
title(u'Transformasjon fra sensormåling til RGB lysintensiter');
```

### Skjæring (slicing) av arrayer

- Ofte vil man ikke plukke ut enkelte elementer, f eks en kolonne
- Dette betegnes som *slicing* (*skjæring*)
- Slicing er veldig nyttig (og meget populært som eksamensoppgave ...)
- Slicing kan anvendes på en- og todimensjonale arrayer
- Slicing anvendes på hver dimensjon for seg i todimensjonale arrayer

#### Regler for slicing

- La `a` være et endimensjonalt array
- La `N = len(a)` være antall elementer i `a`
- La `m` og `n` være heltall med `0 <= m <= n <= N-1`
- La `s` være et heltatll med `1 <= s`

| Slice | Elementer plukket ut | Antall elementer |
| - | - | - |
| `a[m:n]` | `a[m], a[m+1], a[m+2], ..., a[n-1]` | `n-m` |
| `a[:n]` | `a[0], a[1], a[2], ..., a[n-1]` | `N` |
| `a[m:]` | `a[m], a[m+1], a[m+2], ..., a[N-1]` | `N-m` |
| `a[::s]` | `a[0], a[s], a[2*s], ..., a[k*s]` | |
| `a[m:n:s]` | `a[m], a[m+s], a[m+2*s], ..., a[m+j*s]` |  | |

- `a[k*s]` er det siste elementet slik at `k*s <= N-1`
- `a[m+j*s]` er det siste elementet slik at `m+j*s <= n-1`

#### Eksempler for en todimensjonal array

```python
b = array([[11, 12, 13, 14, 15], [21, 22, 23, 24, 25], [31, 32, 33, 34, 35], [41, 42, 43, 44, 45]])
print b
```

```python
print b[:, 1]   # kolonne med indeks 1
```

```python
print b[:, 3]   # kolonne med indeks 3
```

```python
print b[0, :]   # rad med indeks 0
```

```python
print b[1, :]   # rad med indeks 1
```

```python
print b[:, :2]  # de første to kolonnene
```

```python
print b[1:, :]    # alle rader fom andre rad
```

```python
print b[:, ::2]   # annenhver kolonne
```

```python
print b[1:, ::2]  # alle rader fom andre rad, annenhver kolonne
```

### Solinnstråling i løpet av året

- I neste grafikk skal vi se på solinnstråling
- Denne måles som energi innstrålt per areal i løpet av døgnet, enheten er $MJ/m^2$
- Måles både globalt (hele spektrum, kolonne 3 i arrayet) og i UV-området (kolonne 4, prosentandel UV av globalinnstråling)
- Vi ser bare på globalstrålingen 

```python
e_glob = weather[:, 3]
print e_glob.min(), e_glob.max()
```

```python
plot(day, e_glob, 'o');
xlabel(u'Døgn regnet fra nyttår');
ylabel(u'Stråling [MJ/m$^2$]');
title(u'Globalstråling for Ås i 2012');
xlim(0, 367);
ylim(0, 30);
```

#### Analyse av kurven

- Vi ser en klar innhyllende for kurven
- Under den innhyllende spriker verdiene mye
    - Hvorfor ser vi innhyllende?
    - Hvorfor spriker verdien så mye?
- Hypothese:
    - Periodisk endring pga jordens posisjon i forhold til solen
    - Enkleste periodiske funksjon: sinus eller kosinus
    - Maksium rundt sommersolverv 21. juni
    - Matematisk modell
    $$E(d) = a + b \cos\frac{2\pi ( d - d_{max} )}{366}$$
    - med
        - $d$: døgn i året
        - $d_{max}$: døgn med maksimal stråling ($\cos 0=1$)
    - fra grafen ovenfor kan vi avlese
        - $a\approx 14$
        - $b\approx 14$
    - døgnnummer for 21. juni må være litt under 180, vi finner det som

```python
d_max = array([31, 29, 31, 30, 31, 21]).sum()
print d_max
```

- Nå kan vi beregne verdier for vår modell og plotte de sammen med dataene

```python
e_modell_max = 14 + 14 * cos(2 * pi * (day - d_max) / 366.)
plot(day, e_glob, 'o', label=u'Målinger');
plot(day, e_modell_max, '-', linewidth=2, label='Modell');
xlabel(u'Døgn regnet fra nyttår');
ylabel(u'Stråling [MJ/m$^2$]');
title(u'Globalstråling for Ås i 2012');
xlim(0, 367);
ylim(0, 30);
legend();
```

- Gitt at vi har gjettet modellen, passer den forbausende godt
- Passer samme modellen også for den nedre grensen?
- Vi estimerer fra figuren $a'\approx 2$, $b'\approx 2$ for den nedre grensen

```python
e_modell_max = 14 + 14 * cos(2 * pi * (day - d_max) / 366.)
e_modell_min =  2 +  2 * cos(2 * pi * (day - d_max) / 366.)
plot(day, e_glob, 'o', label=u'Målinger');
plot(day, e_modell_max, '-', linewidth=2, label='Modell, max');
plot(day, e_modell_min, '-', linewidth=2, label='Modell, min');
xlabel(u'Døgn regnet fra nyttår');
ylabel(u'Stråling [MJ/m$^2$]');
title(u'Globalstråling for Ås i 2012');
xlim(0, 367);
ylim(0, 30);
legend();
```

- Også dette passer rimelig godt
- Vi kan trekke foreløpige konklusjoner:
    - Den årlige variasjonen i solstrålingen kan beskrives med en sinusfunksjon
    - En tett skydekke reduserer solstrålingen med ca en faktor 7
- Disse foreløpige resultatene må undersøkes videre, f eks ved å se på flere år, andre målinger, grundigere teoretisk analyse, mm

### Korrelasjonsanalyse: Temperatur som funksjon av innstråling

- Vi kan nå plotte for hver dag innstrålingen på x- og temperaturen på y-aksen
- Vil vi se en sammenheng?

```python
plot(e_glob, T_avg, 'o');
xlabel(u'Globalstråling [MJ/m$^2$]');
ylabel(u'Døgnmiddeltemperatur [C]');
title(u'Sammenheng mellom temperatur og stråling på Ås i 2012');
xlim(0, 30);
ylim(-20, 25);
```

- Vi ser at datapunktene for globalstråling $<5 \mathrm{MJ}/\mathrm{m}^2$ nesten ligger på en vertikal linje
    - Det tyder på lite sammenheng mellom stråling og temperatur for disse datapunktene
    - De er stort sett fra midtvinter, når solen er meget svak
- For globalstråling $>5 \mathrm{MJ}/\mathrm{m}^2$ er data også ganske spredt, men vi ser en svak sammenheng: temperaturen øker med strålingen
- Vi kan skille data i punkter med høy og lav stråling og plotte disse i forskjellige farger
    - Til dette benytter vi logisk slicing: `e_glob[e_glob < 5]` er alle `e_glob` verdier mindre enn 5

```python
e_glob_low = e_glob[e_glob < 5]
e_glob_high = e_glob[e_glob >= 5]
T_avg_low = T_avg[e_glob < 5]
T_avg_high = T_avg[e_glob >= 5]

plot(e_glob_low, T_avg_low, 'bo', label=r'$E < 5\mathrm{MJ}/\mathrm{m}^2$');
plot(e_glob_high, T_avg_high, 'ro', label=r'$E \geq 5\mathrm{MJ}/\mathrm{m}^2$');
xlabel(u'Globalstråling [$\mathrm{MJ}/\mathrm{m}^2$]');
ylabel(u'Døgnmiddeltemperatur [C]');
title(u'Sammenheng mellom stråling og temperatur på Ås i 2012');
xlim(0, 30);
ylim(-20, 25);
legend(loc='lower right');
```

- Vi kan nå bruke lineær regresjon for å finne den rette linjen som best beskriver de røde punktene
- Konkret finner vi en funksjon

  $$T(E) = c E + d$$

  slik at den kvadratiske feilen

  $$\Delta = \sum_{j} (T(E_j) - T_j)^2$$

  blir minst mulig. $E_j$ og $T_j$ er strålings- og temperaturverdiene for de røde datapunktene.

```python
c, d = polyfit(e_glob_high, T_avg_high, 1)
print "c = ", c
print "d = ", d
```

- Om funksjonen `polyfit(x_data, y_data, grad)`:
    - Den finner det polynomet av gitt `grad` som gir den minste feilen $\Delta$ mellom polynomet og dataene
    - Vi vil ha en rett linje, dvs et polynom av grad 1
    - Funksjonen gir tilbake stigningen $c$ og y-akseavsnittet $d$
- Nå tegner vi denne linjen sammen med dataene
- Først beregner vi funksjonen for noen datapunkter, slik at vi har noe å plotte

```python
x_e = linspace(5, 30)
T_fit = c * x_e + d

plot(e_glob_low, T_avg_low, 'bo', label=r'$E < 5\mathrm{MJ}/\mathrm{m}^2$');
plot(e_glob_high, T_avg_high, 'ro', label=r'$E \geq 5\mathrm{MJ}/\mathrm{m}^2$');
plot(x_e, T_fit, 'g-', linewidth=2, label=r'$T = 0.44\frac{^\circ\mathrm{C}}{\mathrm{MJ}/\mathrm{m}^2} E + 3.98 ^\circ\mathrm{C}$');
xlabel(u'Globalstråling $E$ [$\mathrm{MJ}/\mathrm{m}^2$]');
ylabel(u'Døgnmiddeltemperatur $T$ [$^\circ$C]');
title(u'Sammenheng mellom temperatur og stråling på Ås i 2012');
xlim(0, 30);
ylim(-20, 25);
legend(loc='lower right');
```

- Den grønne linjen tegnes bare for $E\geq 5$ fordi vi benyttet bare de røde punktene til fitting
- Vi ser en grov sammenheng mellom temperatur og stråling: i den varme årstiden øker temperaturen med $0.44 \frac{^\circ\mathrm{C}}{\mathrm{MJ}/\mathrm{m}^2}$ dvs med snaut et halvt Kelvin per Megajoule stråling per kvadratmeter.

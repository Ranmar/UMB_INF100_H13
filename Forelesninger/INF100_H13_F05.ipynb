### Automatisk oppretting av arrayer

#### Tall med fast intervall: `linspace(a, b)`
- Lager array med 50 tall fra $a$ til $b$
- Både $a$ og $b$ er inkludert
- Fast avstand mellom tallene: $(b-a)/49$
- En array med 100 tall fåes med `linspace(a, b, num=100)`

```python
linspace(1, 50)
```

```python
linspace(1, 50, num=10)
```

#### Tall med logaritmisk intervall: `logspace(a, b)`
- Lager array med 50 tall fra $10^a$ til $10^b$
- Både $10^a$ og $10^b$ er inkludert
- Logaritmisk avstand mellom tallene

```python
logspace(0, 2)
```

### Ikke stol blind på grafikker, husk kritisk vurdering!

#### Hva er galt med denne grafikken?

```python
x = linspace(0, 300)
plot(x, cos(x), '-');
```

- Legge merke til $x$-aksen: den går fra 0 til 300
- Grafen som vises kan altså ikke være en vanlig kosinus: den har periode fra $0$ til $2\pi$
- For å se hva som foregår, lager vi grafen igjen, men med flere punkter langs $x$-aksen

```python
xh = linspace(0, 300, num=2000)
plot(xh, cos(xh), 'g-');
```

- Vi ser nå at vi egentlig har svært mange kosinusperioder i plottet (nøyaktig $300/(2\pi)\approx 48$)
- `linspace(0, 300)` gir oss bare 50 punkter langs $x$-aksen
- Vi har bare tegnet **ett datapunkt per periode** i den første grafen
- For å se dette bedre, plotter vi kosinus igjen med få punkter og med mange punkter, men bare fra 0 til 60

```python
plot(x[x<60], cos(x[x<60]), 'bo-');
plot(xh[xh<60], cos(xh[xh<60]), 'g-');
```

- Den grønne linjen viser den korrekte kosinusfunksjonen
- De blå punktene viser de datapunktene vi benyttet i den første grafen
- Den blå linjen forbinder de blå punktene og er den linjen vi så i den første grafen

### Skjæring (slicing) av arrayer

- Ofte vil man ikke plukke ut enkelte elementer, f eks en kolonne
- Dette betegnes som *slicing* (*skjæring*)
- Slicing er veldig nyttig (og meget populært som eksamensoppgave ...)

#### Slicing av endimensjonale arrayer

- Vi definerer en array til illustrasjon

```python
a = linspace(10, 100, num=10)
print a
```

- `a[m:n]` gir oss et array med elementene 
    - fra og med index `m`
    - til men **uten** index `n`
    - samme logikk som for `range()`: stop-verdien er ikke med
    - totalt antall elementer plukket: `n-m`
    - husk: første element har index 0

```python
print a[2:4]
```

```python
print a[2:7]
```

```python
print a[5:6]
```

- `a[:n]` gir oss de første `n` elementene
    - fra og med index 0
    - til men uten index n

```python
print a[:1]
```

```python
print a[:5]
```

- `a[m:]` gir oss alle elementer fra og med index `m` 

```python
print a[1:]
```

```python
print a[5:]
```

- `a[m:n:s]` gir oss en array med elementene 
    - fra og med indeks `m`
    - til men uten indeks `n`
    - med **skritt** `s`
    - dvs `a[m], a[m+s], a[m+2*s], ...`

```python
print a[1:8:2]
```

```python
print a[2:9:3]
```

- Vi kan sløyfe `m`, `n`, eller begge deler for å begynne med første eller gå til siste element

```python
print a[:5:2]
```

```python
print a[5::2]
```

```python
print a[::2]
```

- `a[:]` gir oss hele arrayen

```python
print a[:]
```

#### Slicing av todimensjonale arrayer

- Samme logikk som for endimensjonale arrayer
- Vi definerer en array til illustrasjon

```python
b = array([[11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34]])
print b
```

- b[:, k] gir oss kolonnen med indeks `k`
    - selv om vi plukker en kolonne, vises den i utskrift som rad: PyLab ser den som endimensjonal array

```python
print b[:, 1]
```

```python
print b[:, 3]
```

- `b[r, :]` gir oss raden med indeks `r`

```python
print b[0, :]
```

```python
print b[1, :]
```

- Hvis vis vil ha flere rader eller kolonner, gjør vi som for endimensjonale arrayer
    - Vi får da en todimensjonal array som resultat

```python
print b[:, :2]    # de første to kolonnene
```

```python
print b[1:, :]    # alle rader fom andre rad
```

```python
print b[:, ::2]   # annenhver kolonne
```

- Vi kan kombinere plukk fra rader og kolonner

```python
print b[1:, ::2]
```

### Innlesing av data til Python

- Vi kan lese data fra tekstfil med `loadtxt()`

```python
weather = loadtxt('Data/weather_umb_2012.csv', delimiter=';', usecols=(1, 2, 3, 4, 5))
```

- Hva betyr denne linjen?
    - Det første argumentet til `loadtxt()` er navnet til filen som skal leses
    - Siden filen ligger i mappen Data som ligger i mappen med denne notebooken, står mappenavnet foran filnavnet
    - `delimiter=';'` forteller `loadtxt()` at kolonnene skilles med semikolon
    - `usecols=(1, 2, 3, 4, 5)` forteller `loadtxt()` at den skal lese data fra kolonnene 1, 2, ..., 5
        - NB: Kolonnene **telles fra 0**, vi hopper altså over den første kolonnen med datoene
        - Vi hopper over denne, fordi verdiene i den første kolonnen ikke er enkle tallverdier
        - Vi ser senere hvordan vi kan håndtere datoverdier
    - Dataene lagres i en array med navnet `weather`
    
- Vi kan se på størrelsen til `weather` og på dataene i de første radene

```python
weather.shape
```

- 366 rader: data for 366 døgn (2012 var skuddår)

### Temperatur i løpet av året

- Som en første plot, ser vi på døgnmiddeltemperatur for hver dag (andre kolonne i filen, første kolonne i arrayet)
- Først finner vi minimums- og maksimumstemperatur

```python
weather[:, 0].min()
```

- Hvis vi bare skriver et uttrykk som på linjen ovenfor, ser vi den "sanne" verdien som er lagret i datamaskinen, som er så vidt over $-14.7$
- Hvis vi bruker `print`, runder Python litt

```python
print weather[:, 0].max()
```

- Med informasjonen ovenfra velger vi y-grensen ved -20 og +25

```python
plot(weather[:, 0], 'o', label=u'Døgnmiddel');
xlabel(u'Døgn regnet fra nyttår');
ylabel(u'Temperatur [C]');
title(u'Temperatur for Ås i 2012');
xlim(0, 367);
ylim(-20, 25);
legend();
```

- Legg merke til at vi ikke har oppgitt noen x-verdier, bare y-verdiene
- `plot()` legger da automatisk tallene 0, 1, ..., 365 til som x-verdier
- **NB:** Hvis tekst inneholder særbokstaver, må vi sette `u` foran det første hermetegn
    - Vi merker da teksten som en *Unicode*-streng
    - Bare slike kan håndtere særbokstaver

- I neste plot skal vi også legge til minimums- og maksimumstemperatur per døgn
- Vi gir de forskjellige data først kortnavn, slik at koden blir mer oversiktelig
- Vi ser på minimal minimums- og maksimal maksimumstemperatur for å finne grensene

```python
T_avg = weather[:, 0]
T_min = weather[:, 1]
T_max = weather[:, 2]
```

```python
print T_min.min()
```

```python
print T_max.max()
```
